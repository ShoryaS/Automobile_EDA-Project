```python
plt.figure(figsize=(18,5))
plt.subplots_adjust(hspace=0.2,wspace=0.5)
plt.subplot(1,4,1)
plt.title('city-mpg')
plt.plot(automobile['city-mpg'],'r')
plt.subplot(1,4,2)
plt.title('horsepower')
plt.plot(automobile.horsepower,'b')
plt.subplot(1,4,3)
plt.title('symboling')
plt.plot(automobile.symboling,'g')
plt.subplot(1,4,4)
plt.title('price')
plt.plot(automobile.price,'y')
plt.show()
```

```python
table=pd.pivot_table(automobile,values='price',index=['num-of-doors','body-style'],columns=['make','drive-wheels','curb-weight','horsepower'])
table
```

```python
automobile[['city-mpg','price']].sort_values('price',ascending=False)
```

```python
# Which manufacturer's cars is the safest based on risk rating

most_safety_car=automobile.groupby('make').agg({'symboling':[np.size,np.mean]})
most_safety_car.sort_values([('symboling','mean')],ascending=False)
```

```python
# Most preferred car based on 'fuel-type',number of doors and body style
most_prefer=pd.pivot_table(automobile,values='price',index=['num-of-doors'],columns=['fuel-type','body-style'])
most_prefer
#most_prefer=automobile.groupby('fuel-type','num-of-doors')
```

```python
preferred=automobile.groupby(['num-of-doors','fuel-type','body-style']).size().sort_values(ascending=False)
preferred
```

```python
# Which car is most priced on an average

avg_price=automobile.groupby('make').agg({'price':[np.size,np.mean]})
avg_price.sort_values([('price','mean')],ascending=False)
```

```python
# Car with highest average price with at least '6' models

at_least=avg_price['price']['size']>=6
avg_price[at_least].sort_values([('price','mean')],ascending=False)
```

```python
#. No cars having 2,3,8,12 cylinders for diesel cars, whereas gas cars comes with every range of cylinders

preferred1=pd.pivot_table(automobile,values='city-mpg',index=['fuel-type'],columns=['num-of-cylinders'])
preferred1
```

```python
# Analysis between make, drive wheels and curb weight
#Cars with more curb weight have 'rear wheel drive' mostly(exception being AUDI which choose to have 4wd instead of rwd
# for higher curb weight cars)

bivariate_ana=pd.pivot_table(automobile,values='curb-weight',index=['make'],columns=['drive-wheels'])
bivariate_ana.sort_values('fwd',ascending=False)
```

```python
trivariate_ana=pd.pivot_table(automobile,values='horsepower',index=['make'],columns=['drive-wheels'])
trivariate_ana.sort_values('fwd',ascending=False)
```

```python
aut1=automobile[['make','city-mpg','price']]
```
